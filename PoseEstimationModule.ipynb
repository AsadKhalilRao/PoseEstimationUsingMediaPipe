To make a module we have to write 
`if ___name__=="__main__":
    main()`
if the whole script or cell is run  then the `main()` function will run 
if we are just calling another function then `if ___name__=="__main__":main()` will not run
Then we write 
`def main():`Whenever we want to see what a module is capable of or a testing script we can put it in the main function
`class poseDetector():`to create objects and methods 

```python
import cv2
import mediapipe as mp
import time

class poseDetector():
    def __init__(self,mode=False,model_complexity=1,smooth=True,detectionCon=0.5,trackCon=0.5):
        # Whenever we create a new object it will have it's own variables
        # self.something is the varaible of the newly created object
        # Whenever we are using varaible with in the class we will write self.something
        #self.something is the specific varaible of our object is the one that use has provided
        self.mode=mode
        self.model_complexity=model_complexity
        self.smooth=smooth
        self.detectionCon=detectionCon
        self.trackCon=trackCon
        
        #Step#1
        #Creating our Model to detect our pose
        self.mpDraw=mp.solutions.drawing_utils
        self.mpPose=mp.solutions.pose
        self.pose=self.mpPose.Pose(static_image_mode=self.mode,model_complexity=self.model_complexity,smooth_landmarks=self.smooth,
                                   min_detection_confidence=self.detectionCon,min_tracking_confidence=self.trackCon)
      
    def findPose(self,img,draw=True):
        
        #Converting BGR image to RGB
        imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        self.results=self.pose.process(imgRGB)
        if self.results.pose_landmarks:
            if draw:
                #Getting actual land marks
                # print("Actual Land marks",results.pose_landmarks)
                #Drawing Land marks and connections between the landmarks
                #sending in the image and land marks (results.pose_landmarks)
                self.mpDraw.draw_landmarks(img,self.results.pose_landmarks,self.mpPose.POSE_CONNECTIONS)
        return img
            
    def findPosition(self,img,draw=True):
        
        
        landmarkList=[]
        if self.results.pose_landmarks:
            # Extracting the information from results.pose_landmarks by looing though the results.pose_landmarks
            for id_of_landmark,landmark in enumerate(self.results.pose_landmarks.landmark):    
                # print("\nID of landmark \n",id_of_landmark,"\nLandmark\n", landmark)
                #Getting the values x,y,z of land mark in pixels
                #Getting the dimension of image
                height_of_image,width_of_image,color_channels_of_image=img.shape
                cx,cy=int(landmark.x*width_of_image),int(landmark.y*height_of_image)
                landmarkList.append([id_of_landmark, cx,cy])
                if draw:
                    #Drawing circle on the cx,cy pixel positions
                    cv2.circle(img,(cx,cy),9,(0,255,0),cv2.FILLED)
        return landmarkList
def main():
    cap=cv2.VideoCapture("PoseVideos/pose1.mp4")
    if not cap.isOpened():
        print("Error: Could not open video.")
        return
    previousTime=0
    detector=poseDetector()
    while True:
        # img:It is in BGR
        success,img=cap.read()
        if not success:
            break
        img=detector.findPose(img)
        lmList=detector.findPosition(img,draw=False)
        if len(lmList)!=0
            # print("Land mark list",lmList)
            # Tracking particular Landmark only
            #Change 14 ,any number from 0-32
            landmarkID=int(32)
            specifLandmark=lmList[landmarkID]
            print("Specif Land Mark",specifLandmark)
            x_Position_Of_Specif_Landmark=specifLandmark[1]
            y_Position_Of_Specif_Landmark=specifLandmark[2]
            # Drawing Specific Land Mark
            cv2.circle(img,(x_Position_Of_Specif_Landmark,y_Position_Of_Specif_Landmark),9,(0,255,0),cv2.FILLED)
        currentTime=time.time()
        if previousTime != 0:
            fps=1/(currentTime-previousTime)
            cv2.putText(img,str(int(fps)),(70,50),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),3)
        previousTime=currentTime
        cv2.imshow("Image",img)
        
        
    
        if cv2.waitKey(1) & 0xFF == 27:
            break
    cap.release()
    cv2.destroyAllWindows()

if __name__== "__main__":
    try:
        main()
    finally:
        cv2.destroyAllWindows()
        
```
