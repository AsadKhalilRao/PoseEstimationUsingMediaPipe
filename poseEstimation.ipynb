```python
import cv2
import mediapipe as mp
cap=cv2.VideoCapture("PoseVideos/pose1.mp4")
import time
mpDraw=mp.solutions.drawing_utils
#Step#1
#Creating our Model to detect our pose
mpPose=mp.solutions.pose
#Step#2
# Creating an Object pose=mpPose.Pose()
#Step#3
#Giving in the parameters to the Object pose=mpPose.Pose()
#Parameters:
#1)static_image_mode=True will always detect based on the model,always try to find the new detections
#2)static_image_mode=False will try to detect and when the confidence is high it will keep tracking
pose=mpPose.Pose()
previousTime=0
while True:
    # img:It is in BGR
    success,img=cap.read()
    if not success:
        break   
    #Step#4
    #Converting BGR image to RGB
    imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    #Step#5
    #Sending RGB img to Model
    results=pose.process(imgRGB)
    
    #Step#6
    #Getting actual land marks
    # print("Actual Land marks",results.pose_landmarks)
    
    #Step#7
    #Checking whether land mark is detected or NOT
    if results.pose_landmarks:
    
    #Step#8
    #Drawing Land marks
    #sending in the image and land marks (results.pose_landmarks)    
        # mpDraw.draw_landmarks(img,results.pose_landmarks)
    
    
    #Step#9
    #Drawing Connections /Lines between Land marks or Points
    #sending in the image and land marks (results.pose_landmarks)    
        mpDraw.draw_landmarks(img,results.pose_landmarks,mpPose.POSE_CONNECTIONS)
    #Step 10
    # Extracting the information from results.pose_landmarks by looing though the results.pose_landmarks
        for id_of_landmark,landmark in enumerate(results.pose_landmarks.landmark):
            
            # print("\nID of landmark \n",id_of_landmark,"\nLandmark\n", landmark)
            #Step#9 Getting the values x,y,z of land mark in pixels
            #Getting the dimension of image
            height_of_image,width_of_image,color_channels_of_image=img.shape
            cx,cy=int(landmark.x*width_of_image),int(landmark.y*height_of_image)
            #Step#9 Drawing circle on the cx,cy pixel positions
            cv2.circle(img,(cx,cy))
            
            
            
       
    currentTime=time.time()
    fps=1/(currentTime-previousTime)
    previousTime=currentTime
    cv2.putText(img,str(int(fps)),(70,50),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),3)
    cv2.imshow("Image",img)

    if cv2.waitKey(1) & 0xFF == 27:
        break
    
cap.release()
cv2.destroyAllWindows()
```
